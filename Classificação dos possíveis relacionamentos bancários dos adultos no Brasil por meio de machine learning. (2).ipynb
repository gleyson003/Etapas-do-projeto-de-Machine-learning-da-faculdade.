##### Transformação de dados

De acordo com o Blog Kondado: "Transformação de dados é o processo de converter a estrutura ou o formato dos seus dados em outro formato ou estrutura em que seja possível analisá-los". Como iremos utilizar o método Naive Bayes, é necessário que os dados sejam transformados para o formato de número inteiro. Então, os dados da coluna "Região" chamados Norte, Nordeste, Centro-oeste, sudeste e sul serão substituidos respectivamente por 0,1,2,3 e 4.

```python
data['Região'].replace({'Norte':0, 'Nordeste':1, 'Centro-oeste':2, 'Sudeste':3, 'Sul':4}, inplace=True)

data['Tem conta'].replace({'s':1, 'n':0}, inplace=True)
data.head()
```

Após realizado a normalização da coluna "Região", iremos criar uma nova coluna chamada "Relacionamentos", onde essa terá os valores de uma variável chamada "sr". Essa variável receberá os valores de uma operação lógica, em seguida mudaremos o tipo dos dados da coluna de bool para int. Os valores dessa nova coluna serão os resultados que queremos obter na aplicação dos nossos algoritmos. As amostras com idade igual ou maior que 15 são um possível relacionamento, ou seja, o valor "1" significa um posível relacionamento e o "0" não pode ser um relacionamento.

```python
sr = data['Idade'] >= 15
data['Relacionamento'] = sr
data['Relacionamento'] = data['Relacionamento'].astype(int)
data.head()
```
